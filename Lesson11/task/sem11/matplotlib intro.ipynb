### Задание 4
Построить графики функции из таблицы в пределах 1 < x < 1000



```python
def constant(x, c):
    return [c for _ in x]


def lenear(x):
    return [n for n in x]


def logarithmic(x):
    return [math.log(n) for n in x]


x = [x for x in range(1, 1000)]

plt.plot(x, constant(x, 20))
plt.plot(x, lenear(x))
plt.plot(x, logarithmic(x))

plt.ylim(0, 100)
plt.xlim(0, 100)
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

### Задание 5
Отобразите несколько графиков функций из прошлого (4-го)
- задайте тип линии
- задайте цвета первым трем функциям: red, green, blue
- задайте разные маркеры нескольким функциям


```python
def constant(x, c):
    return [c for _ in x]


def lenear(x):
    return [n for n in x]


def logarithmic(x):
    return [math.log(n) for n in x]


x = [x for x in range(1, 1000)]

points = [
    x, constant(x, 20),
    x, lenear(x),
    x, logarithmic(x)
]

plt.plot(*points)

plt.ylim(0, 100)
plt.xlim(0, 100)
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

### Задание 6
Залейте цветом между функциями y = x и y = x * 2


```python
x = range(0, 20)
y = [n for n in x]
y2 = [n*2 for n in x]

points = [
    x, y,
    x, y2
]

plt.plot(*points)

plt.fill_between(x, y2, x, y, color="green", interpolate=True)


plt.show()
```
